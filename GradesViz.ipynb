#**Subjects Section**

###Minor Subjects

Retrieving subject rows that are GE, ASF, and PE

```python
minor_subjects = dataset[dataset['subj_no'].str.startswith(('GE', 'PE', 'ASF'))]
minor_subjects
```

Retrieving the value count of GE subjects - True Values are the GE values

```python
minor_subjects['subj_no'].str.startswith('GE').value_counts()
```

###Major Subjects

Retrieving subject rows that are DSC, CS, and Math

```python
major_subjects = dataset[dataset['subj_no'].str.startswith(('DSC', 'CS', 'MATH'))]
major_subjects
```

Retrieving the value count of DSC subjects - True Values are the DSC values

```python
major_subjects['subj_no'].str.startswith('DSC').value_counts()
```

Retrieving the value count of each subject by splitting subj_no

```python
subject_prefixes = ['GE', 'PE', 'ASF', 'DSC', 'CS', 'MATH']
prefixes = dataset['subj_no'].str.split().str[0]
subject_counts = prefixes[prefixes.isin(subject_prefixes)].value_counts()
```

Pie chart displaying the percentages of all GE, PE, ASF, DSC, CS and Math subjects

```python
plt.figure(figsize=(10, 8))
plt.pie(subject_counts, labels=subject_counts.index, autopct='%1.1f%%')
plt.title('Percentage of GE, PE, ASF, DSC, CS, and Math Subjects')
plt.show()
```
